# GCaMP + jRGECO(1) spectrum

```python
import pandas as pd

input_path = "/Users/rnlghdb/NDL/data/GCaMP + jRGECO(1)/Flna29_gr25uw_1_Subt2__0__81198.txt"
output_path = "/Users/rnlghdb/NDL/data/GCaMP + jRGECO(1)/Flna29_gr25uw_1_Subt2__0__81198.csv"

with open(input_path, "r") as f:
    lines = f.readlines()

# Spectral data 시작 위치 찾기
start_idx = None
for i, line in enumerate(lines):
    if ">>>>>Begin Spectral Data<<<<<" in line:
        start_idx = i + 1
        break

data_lines = [line.strip().split() for line in lines[start_idx:] if line.strip()]

# 첫 번째 2136개 숫자 행 = wavelength
wavelengths = data_lines[0]

# 2139개짜리 행만 실제 timepoint 데이터로 사용
# 앞 3개 = Date, Time, Timestamp
rows = []
for row in data_lines[1:]:
    if len(row) == 2139:
        rows.append(row)

meta = [r[:3] for r in rows]
intensity = [r[3:] for r in rows]

df_intensity = pd.DataFrame(intensity, columns=wavelengths)
df_meta = pd.DataFrame(meta, columns=["Date", "Time", "Timestamp"])

df = pd.concat([df_meta, df_intensity], axis=1)

# 숫자형 변환
df.iloc[:, 2:] = df.iloc[:, 2:].apply(pd.to_numeric)

df.to_csv(output_path, index=False)

print(df.shape)
print(df.head())
print(f"저장 완료: {output_path}")
```

# GCaMP + jRGECO(2) spectrum

```python
import os
import pandas as pd

base_dir = "/Users/rnlghdb/NDL/data/GCaMP + jRGECO(2)"

files = [
    "Flna36_GR25_Trig1_Subt2__0__72440.txt",
    "Flna36_GR25_Trig2_Subt2__0__76959.txt",
    "Flna36_GR25_Trig3_Subt2__0__79786.txt",
    "Flna36_GR25_Trig4_Subt2__0__83650.txt",
    "Flna36_GR25_Trig5_Subt2__0__86908.txt",
    "Flna36_GR25_Trig6_Subt2__0__90634.txt",
    "Flna36_GR25_Trig7_Subt2__0__94297.txt",
    "Flna36_GR25_Trig8_Subt2__0__97790.txt",
    "Flna36_GR25_Trig9_Subt2__0__100777.txt",
    "Flna36_GR25_Trig10_Subt2__0__104370.txt"
]

for filename in files:

    input_path = os.path.join(base_dir, filename)
    output_path = os.path.join(
        base_dir,
        filename.replace(".txt", ".csv")
    )

    with open(input_path, "r") as f:
        lines = f.readlines()

    # Spectral data 시작 위치 찾기
    start_idx = None
    for i, line in enumerate(lines):
        if ">>>>>Begin Spectral Data<<<<<" in line:
            start_idx = i + 1
            break

    data_lines = [
        line.strip().split()
        for line in lines[start_idx:]
        if line.strip()
    ]

    # wavelength row
    wavelengths = data_lines[0]

    # timepoint rows
    rows = []
    for row in data_lines[1:]:
        if len(row) == 2139:
            rows.append(row)

    meta = [r[:3] for r in rows]
    intensity = [r[3:] for r in rows]

    df_intensity = pd.DataFrame(
        intensity,
        columns=wavelengths
    )

    df_meta = pd.DataFrame(
        meta,
        columns=["Date", "Time", "Timestamp"]
    )

    df = pd.concat(
        [df_meta, df_intensity],
        axis=1
    )

    df.iloc[:, 2:] = df.iloc[:, 2:].apply(
        pd.to_numeric,
        errors="coerce"
    )

    df.to_csv(
        output_path,
        index=False
    )

    print(
        f"완료: {filename} -> "
        f"{os.path.basename(output_path)} "
        f"{df.shape}"
    )

print("모든 파일 변환 완료")
```
