```python
import pandas as pd

input_path = "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_1_Subt2__0__81198.csv"
output_path = "/Users/rnlghdb/NDL/data/unmixing_spectrum/Flna29_gr25uw_1_Subt2__0__81198.csv"

df = pd.read_csv(input_path)

# wavelength 추출
wavelengths = df.columns[3:].astype(float)

# intensity matrix만 추출
spectra = df.iloc[:, 3:].apply(pd.to_numeric, errors="coerce")

# 450–700 nm만 사용
mask = (wavelengths >= 450) & (wavelengths <= 700)

mixed = spectra.loc[:, df.columns[3:][mask]]

# index를 Time1, Time2...로 변경
mixed.index = [f"Time{i+1}" for i in range(len(mixed))]

# 저장
mixed.to_csv(output_path)

print(mixed.shape)
print(mixed.head())
print(f"저장 완료: {output_path}")
```

```python
import os
import pandas as pd

input_files = [
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_2_Subt2__0__81198.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_3_Subt2__0__88268.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_4_Subt2__0__92821.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_5_Subt2__0__96263.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_6_Subt2__0__99471.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_7_Subt2__0__102569.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_8_Subt2__0__107144.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_9_Subt2__0__111258.csv",
    "/Users/rnlghdb/NDL/data/GCaMP+jRGECO(1)/Flna29_gr25uw_10_Subt2__0__115761.csv"
]

output_dir = "/Users/rnlghdb/NDL/data/unmixing_spectrum"
os.makedirs(output_dir, exist_ok=True)

for input_path in input_files:

    df = pd.read_csv(input_path)

    # wavelength 추출
    wavelengths = df.columns[3:].astype(float)

    # intensity matrix
    spectra = df.iloc[:, 3:].apply(
        pd.to_numeric,
        errors="coerce"
    )

    # 450–700 nm
    mask = (
        (wavelengths >= 450) &
        (wavelengths <= 700)
    )

    mixed = spectra.loc[
        :,
        df.columns[3:][mask]
    ]

    # Time1 ~ TimeN
    mixed.index = [
        f"Time{i+1}"
        for i in range(len(mixed))
    ]

    # 파일명 유지
    output_path = os.path.join(
        output_dir,
        os.path.basename(input_path)
    )

    mixed.to_csv(output_path)

    print(
        f"완료: {os.path.basename(input_path)} "
        f"{mixed.shape}"
    )

print("모든 파일 저장 완료")
```
